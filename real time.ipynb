```python
import cv2
import numpy as np
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model
```

```python
# Load the full model
loaded_model = load_model("emotion_detector.h5")
# Define the labels
labels = ['sad', 'fear', 'happy', 'neutral', 'surprise', 'angry', 'disgust']
# Load Haar cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
def predict_expression(frame, model):
    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(48, 48))

    for (x, y, w, h) in faces:
        # Extract the region of interest (face)
        roi = gray_frame[y:y+h, x:x+w]

        # Preprocess the face image
        roi_resized = cv2.resize(roi, (48, 48))
        roi_normalized = roi_resized / 255.0
        roi_array = img_to_array(roi_normalized)
        roi_array = np.expand_dims(roi_array, axis=0)

        # Predict the expression
        prediction = model.predict(roi_array)
        predicted_label = labels[np.argmax(prediction)]

        # Draw a rectangle around the face and label the expression
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        cv2.putText(frame, predicted_label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    return frame
```

```python
cap = cv2.VideoCapture(0)  

print("Starting real-time face expression detection. Press 'q' to exit.")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Detect and display the results
    frame_with_detections = predict_expression(frame, loaded_model)
    cv2.imshow('Face Expression Recognition', frame_with_detections)

    # Break the loop with 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close windows
cap.release()
cv2.destroyAllWindows()
```
